Custom queue implementation using SLL

```python
#Time complexity of all operations: O(1)
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class myQueue:
    def __init__(self):
        self.front = None
        self.rear = None
        self.numNodes = 0
        
    def enqueue(self, data):
        temp = Node(data)
        if self.rear == None:
            self.rear = temp
            self.front = temp
        else:
            self.rear.next = temp
            self.rear = temp
        self.numNodes += 1
        
    def dequeue(self):
        if self.front == None:
            return None
        else:
            dequeuedItem = self.front.data
            self.front = self.front.next
            if self.front == None:
                self.rear = None
            self.numNodes -= 1
            return dequeuedItem
        
    def getRear(self):
        if self.rear == None:
            return None
        else:
            return self.rear.data
    
    def getFront(self):
        if self.front == None:
            return None
        else:
            return self.front.data
        
    def size(self):
        return self.numNodes
    
    def isEmpty(self):
        if self.numNodes == 0:
            return True
        else:
            return False
        
q = myQueue()
q.enqueue(10)
q.enqueue(20)
q.enqueue(30)
print(f"Size: {q.size()}")
print(f"Rear: {q.getRear()}")
print(f"Front: {q.getFront()}")
print(f"Dequeued Element: {q.dequeue()}")
print(f"Size: {q.size()}")
q.enqueue(40)
print(f"Is empty?: {q.isEmpty()}")
print(f"Rear: {q.getRear()}")
print(f"Front: {q.getFront()}")
```

Circular Queue implemetation using list

```python
class myCircularQueue:
    def __init__(self, capacity):
        self.queue = [None]*capacity
        self.capacity = capacity
        self.front = 0
        self.rear = 0
        self.size = 0
        
    def enque(self, data):
        if self.size == self.capacity:
            print("Overflow")
            return -1
        
        else:
            self.queue[self.rear] = data
            self.rear = (self.rear+1)%self.capacity
            self.size += 1
            return 1
    
    def deque(self):
        if self.size == 0:
            print("underflow")
            return -1
        
        else:
            dequedItem = self.queue[self.front]
            self.queue[self.front] = None
            self.front = (self.front+1)%self.capacity
            self.size -= 1
            return dequedItem
            
    def getRear(self):
        return self.queue[self.rear]
    
    def getFront(self):
        return self.queue[self.front]
    
    def getSize(self):
        return self.size
    
    def isEmpty(self):
        return (self.size == 0)
    
    def printQueue(self):
        print(self.queue)

Q = myCircularQueue(3)
print(f"Rear: {Q.getRear()}")
print(f"Front: {Q.getFront()}")
Q.enque(10)
Q.enque(20)
print(f"Size: {Q.getSize()}")
Q.enque(30)
Q.enque(40)
Q.printQueue()
print(f"Rear: {Q.getRear()}")
print(f"Front: {Q.getFront()}")
print(f"Dequeued Element: {Q.deque()}")
Q.printQueue()
print(f"Size: {Q.getSize()}")
Q.enque(40)
Q.printQueue()
print(f"Is empty?: {Q.isEmpty()}")
print(f"Rear: {Q.getRear()}")
print(f"Front: {Q.getFront()}")
```

Stack implementation using Queue

```python
class Stack:
    def __init__(self):
        self.Q1 = []
        self.Q2 = []
        self.Q1Size = 0
        self.Q2Size = 0
    
    def push(self, data):
        self.Q1.append(data)
        self.Q1Size += 1
    
    def pop(self):
        if self.Q1Size == 0 and self.Q2Size == 0:
            print("Underflow")
            return -1
             
        elif self.Q1Size != 0:
            while self.Q1Size != 1:
                self.Q2.append(self.Q1.pop(0))
                self.Q2Size += 1
                self.Q1Size -= 1
            self.Q1Size -= 1
            return self.Q1.pop(0)
        
        else:
            while self.Q2Size != 1:
                self.Q1.append(self.Q2.pop(0))
                self.Q1Size += 1
                self.Q2Size -= 1
            self.Q2Size -= 1
            return self.Q2.pop(0)
    
    def top(self):
        if self.Q1Size == 0 and self.Q2Size == 0:
            print("Underflow")
            return -1
        
        elif self.Q1Size != 0:
            return self.Q1[-1]
        
        else:
            return self.Q2[-1]
        
    def size(self):
        if self.Q1Size == 0 and self.Q2Size == 0:
            return 0
        
        elif self.Q1Size != 0:
            return self.Q1Size
        
        else:
            return self.Q2Size
        
stk = Stack()
stk.push(10)
stk.push(20)
stk.push(30)
print(f"Size: {stk.size()}")
print(f"Top: {stk.top()}")
print(f"Popped Element: {stk.pop()}")
print(f"Popped Element: {stk.pop()}")
print(f"Size: {stk.size()}")
stk.push(40)
print(f"Top: {stk.top()}")
print(f"Size: {stk.size()}")
print(f"Popped Element: {stk.pop()}")
print(f"Popped Element: {stk.pop()}")
print(f"Popped Element: {stk.pop()}")
```
